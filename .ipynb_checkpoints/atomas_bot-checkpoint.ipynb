```python
import random

class Game:
    # positive -> element number
    # -1 -> plus atom
    # -2 -> minus
    # -3 -> dark plus
    # -4 -> neutrino
    
    def __init__(self):
        self.field = []
        for i in range(6):
            self.field.append(random.randint(1, 3))
        self.spawned_atoms = []
        debug = True
        
    def dprint(self, *arg):
        if (self.debug): print(arg)
        
    def get_atom(self, ind, field = None):
        if field is None: field = self.field
        return field[ind % len(self.field)]
    
    def place_atom(self, ind, num, field = None):
        '''
        Returns copy of new field after placement
        '''
        if field is None: field = self.field
        curr_field = field[:]
        curr_field.insert(ind % len(curr_field), num)
        return curr_field
    
    def remove_atom(self, ind, field = None):
        '''
        Returns copy of new field after removal
        '''
        if field is None: field = self.field
        curr_field = field[:]
        curr_field.pop(ind % len(curr_field))
        return curr_field

    def reaction(self, ind, field = None):
        '''
        Returns copy of new field after reaction
        '''
        if field is None: field = self.field
        curr_field = field[:]
        while len(curr_field) >= 2:
            if (self.get_atom(ind - 1, curr_field) == self.get_atom(ind + 1, curr_field) or self.get_atom(ind, curr_field) == -3):
                new_atom = max(self.get_atom(ind, curr_field) + 1, self.get_atom(ind + 1, curr_field) + 2)
                
                if (self.get_atom(ind, curr_field) == -3):
                    new_atom = max(self.get_atom(ind - 1, curr_field) + 3, self.get_atom(ind + 1, curr_field) + 3)
                    
                curr_field = self.remove_atom(ind + 1, curr_field)
                curr_field = self.remove_atom(ind, curr_field)
                curr_field = self.remove_atom(ind - 1, curr_field)
                
                self.place_atom(ind - 1, new_atom, curr_field)
                ind -= 1
            else:
                break
        return curr_field
    
    def reduce(self, field = None):
        '''
        Returns copy of new field after reduction
        '''
        if field is None: field = self.field
        curr_field = field.copy()
        while True:
            if len(curr_field) <= 2: return
            for i in range(len(curr_field)):
                if ((self.get_atom(i) >= 1 and self.get_atom(i) == self.get_atom(i + 2) and self.get_atom(i + 1) == -1) or self.get_atom(i + 1) == -3):
                    curr_field = self.reaction(i + 1, curr_field)
                break
            else:
                break
        return curr_field
    
                           
    def spawn_atom(self):
        turns = len(self.spawned_atoms)
        l, r = turns // 40 + 1, (turns // 40) * 5 // 4 + 3
        spawned_atom = 0
        if (len(self.spawned_atoms) >= 5 and self.spawned_atoms[-5:].count(-1) == 0):
            spawned_atom = -1
        elif (len(self.spawned_atoms) >= 20 and self.spawned_atoms[-20:].count(-2) == 0):
            spawned_atom = -2
        elif random.randint(1, 60) == 1:
            spawned_atom = -3
        elif random.randint(1, 90) == 1:
            spawned_atom = -4
        else:
            out_of_range = set()
            for atom in self.field:
                if atom < l or atom > r:
                    out_of_range.add(atom)
            pick = random.randint(0, len(self.field) - 1)
            if (pick < len(out_of_range)):
                pick_list = list(out_of_range)
                spawned_atom = pick_list[pick]
            else:
                spawned_atom = random.randint(l, r)
        self.spawned_atoms.append(spawned_atom)
        self.dprint("Spawned atom:", spawned_atom)
        return spawned_atom
                           
    def turn(self):
        spawned_atom = self.spawn_atom()
        self.process_spawn(spawned_atom, False)
    
    def process_spawn(self, spawned_atom, op):
        print(spawned_atom, op)
        if (spawned_atom == -2):
            inp = self.get_input()
            atom = get_atom(inp)
            remove_atom(inp)
            process_spawn(atom, True)
        elif (spawned_atom == -4):
            inp = self.get_input()
            atom = get_atom(inp)
            process_spawn(atom, False)
        elif (op):
            inp = self.get_input()
            process_spawn(spawned_atom if inp == 0 else -1, False)
        else:
            inp = self.get_input()
            self.place_atom(inp, spawned_atom)
            self.field = self.reduce()
                           
    def get_input(self):
        return int(input())
                           
    def play(self):
        print(self.field)
        while (len(self.field) < 19):
            print(self.field)
            self.turn()

game = Game()
game.play()

        
    
```
